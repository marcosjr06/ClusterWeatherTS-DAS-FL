```python
# 1.c – Normalize data from 2021 and 2022 (different headers on .csv)
# SPLIT DATE TIME
# "Date Time" (M/D/YYY HH:MM:SS AM) to Date (YYYY-MM-DD) and Time (HH:MM:SS)

from datetime import datetime


for item in newDFs:
    #create 2 columns: Date and Time
    item.insert(loc = 2, column = 'Date', value = "")
    item.insert(loc = 3, column = 'Time', value = "")

    for index, row in item.iterrows():
        try: # AM/PM
            dateStr = datetime.strptime(row['Date Time'], '%m/%d/%Y %I:%M:%S %p')
            item.at[index,'Date'] = dateStr.date()
            item.at[index,'Time'] = dateStr.time()
        except ValueError:
            try: # No Time
                dateStr = datetime.strptime(row['Date Time'], '%m/%d/%Y')
                item.at[index,'Date'] = dateStr.date()
                item.at[index,'Time'] = "00:00:00"

            except ValueError:
                dateStr = datetime.strptime(row['Date Time'], '%Y-%m-%d %H:%M:%S')
                item.at[index,'Date'] = dateStr.date()
                item.at[index,'Time'] = dateStr.time()

    
    # delete column "Date Time"
    item.drop(['Date Time'], axis=1, inplace=True)
    

    
```
