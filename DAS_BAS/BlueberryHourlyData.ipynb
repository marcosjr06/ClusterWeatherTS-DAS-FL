```python
# Read each CSV file into DataFrame
import pandas as pd

big_df = pd.DataFrame()
dfs = []

for folder in csv_list:
    for file in folder:
        df = pd.read_csv(file)
        dfs.append(df)

big_df = pd.concat(dfs, ignore_index=True)
```

```python
big_df.head()
```

```python
# get data only after 2005
df = big_df[~(big_df['Date'] < '2005-01-01')]

# delete between june and november, every year
for i in range(2005,2024):
    from_ts = str(i)+'-06-01'
    to_ts = str(i)+'-11-30'
    df = df[(df['Date'] < from_ts) | (df['Date'] > to_ts)]


#delete the start of 2005
from_ts = '2005-01-01'
to_ts = '2005-11-30'
df = df[(df['Date'] < from_ts) | (df['Date'] > to_ts)]

#delete the end of 2023
from_ts = '2023-03-01'
to_ts = '2023-12-31'
df = df[(df['Date'] < from_ts) | (df['Date'] > to_ts)]


df.head()
```

```python
df.shape
```

```python
df.columns = ['StationID', 'StationName', 'Date', 'Time', 'SoilTemp_C','Temp60cm_C', 'Temp2m_C','Temp10m_C', 'RelatHumidity','DewPointTemp_C', 'Rainfall_IN','WindSpeed_MPH', 'WindDirection_DEG','SolarRad_WM2']
df.head()
```

```python
dfFiltered = df[['StationName','Date','Time','Temp60cm_C','RelatHumidity']]
dfFiltered.shape
```

```python
# Maximum of measurements
# t = measurements per day
# d = days per season
# y = number of years
# s = number of stations
# MaxMeas = t * d * y * s
# minus 7 years missing (Dade City), minus 30 days (Jay, december/05)
maxMeasurements = (31*96*13) + ((31+28)*96*15) + (182*96*17*15) - (182*96*6*1)

realMeasurements = len(dfFiltered.index)

missingValues = maxMeasurements - realMeasurements


print("Maximum number of measurements: "+str(maxMeasurements))
print("Real number of measurements: "+str(realMeasurements)+" ({:.4f}%)".format(realMeasurements/maxMeasurements*100))
print("Missing measurements: "+str(missingValues)+" ({:.4f}%)".format(missingValues/maxMeasurements*100))
```

```python
Tok = 0
Tnull = 0
RHok = 0
RHnull = 0

for index, row in dfFiltered.iterrows():
    t = row['Temp60cm_C']
    if (pd.isna(t)):
        Tnull += 1
    else:
        Tok +=1
        
    rh = row['RelatHumidity']
    if ((rh>= 0) and (rh<= 100)):
        RHok += 1
    else:
        RHnull +=1
    

print("T Null:" + str(Tnull)+" ({:.4f}%)".format(Tnull/(Tnull+Tok)*100))
print("T Ok:" + str(Tok)+" ({:.4f}%)".format(Tok/(Tnull+Tok)*100))

print("RH Null:" + str(RHnull)+" ({:.4f}%)".format(RHnull/(RHnull+RHok)*100))
print("RH Ok:" + str(RHok)+" ({:.4f}%)".format(RHok/(RHnull+RHok)*100))

print("Total:"+str(Tnull+RHnull))
```

```python
dfSorted = dfCleaned.sort_values(by=['StationName','Date','Time'])
dfSorted.head()
```

```python
# Date: from object (Series) to date (YYYY-MM-DD)
# DOY: from date (YYYY-MM-DD) to day of year (hod = 1-366)

from datetime import datetime
import time

dfSorted['doy']=0
dfSorted['hod'] = 0

for index, row in dfSorted.iterrows():
    strDate = row['Date']
    objDate = datetime.strptime(strDate, "%Y-%m-%d")
    day = objDate.timetuple().tm_yday
    year = objDate.timetuple().tm_year
    mon = objDate.timetuple().tm_mon
    if (year == 2000 or year == 2004 or year == 2008 or 
        year == 2012 or year == 2016 or year == 2020):
        if (mon > 2):
            day = day - 1
    
    dfSorted.at[index,'doy'] = day
    
    strTime = row['Time']
    objTime = time.strptime(strTime, "%H:%M:%S")
    dfSorted.at[index,'hod'] = objTime.tm_hour
```

```python
from tslearn.clustering import TimeSeriesKMeans

Sum_of_squared_distances = []
K = range(1,14)
for k in K:
    km = TimeSeriesKMeans(n_clusters=k, metric="dtw")
    km = km.fit(all_arrays)
    Sum_of_squared_distances.append(km.inertia_)
    
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method For Optimal k - DTW')
plt.show()
```
