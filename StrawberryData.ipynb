```python
import pandas as pd
import os
import glob
  
  
# use glob to get all the csv files 
# in the folder
pathApo = "./Apopka"
pathArc = "./Arcadia"
pathBal = "./Balm"
pathBro = "./Bronson"
pathCit = "./Citra"
pathDov = "./Dover"
pathFlo = "./FloralCity"
pathLak = "./LakeAlfred"
pathPla = "./PlantCity"
pathUma = "./Umatilla"


csv_filesApo = glob.glob(os.path.join(pathApo, "*.csv"))
csv_filesApo.remove("./Apopka/2021.csv")
csv_filesApo.remove("./Apopka/2022.csv")
csv_filesApo.remove("./Apopka/2023.csv")

csv_filesArc = glob.glob(os.path.join(pathArc, "*.csv"))
csv_filesArc.remove("./Arcadia/2021.csv")
csv_filesArc.remove("./Arcadia/2022.csv")
csv_filesArc.remove("./Arcadia/2023.csv")

csv_filesBal = glob.glob(os.path.join(pathBal, "*.csv"))
csv_filesBal.remove("./Balm/2021.csv")
csv_filesBal.remove("./Balm/2022.csv")
csv_filesBal.remove("./Balm/2023.csv")

csv_filesBro = glob.glob(os.path.join(pathBro, "*.csv"))
csv_filesBro.remove("./Bronson/2021.csv")
csv_filesBro.remove("./Bronson/2022.csv")
csv_filesBro.remove("./Bronson/2023.csv")

csv_filesCit = glob.glob(os.path.join(pathCit, "*.csv"))
csv_filesCit.remove("./Citra/2021.csv")
csv_filesCit.remove("./Citra/2022.csv")
csv_filesCit.remove("./Citra/2023.csv")

csv_filesDov = glob.glob(os.path.join(pathDov, "*.csv"))
csv_filesDov.remove("./Dover/2021.csv")
csv_filesDov.remove("./Dover/2022.csv")
csv_filesDov.remove("./Dover/2023.csv")

csv_filesFlo = glob.glob(os.path.join(pathFlo, "*.csv"))
csv_filesFlo.remove("./FloralCity/2022.csv")
csv_filesFlo.remove("./FloralCity/2023.csv")

csv_filesLak = glob.glob(os.path.join(pathLak, "*.csv"))
csv_filesLak.remove("./LakeAlfred/2021.csv")
csv_filesLak.remove("./LakeAlfred/2022.csv")
csv_filesLak.remove("./LakeAlfred/2023.csv")

csv_filesPla = glob.glob(os.path.join(pathPla, "*.csv"))
csv_filesPla.remove("./PlantCity/2022.csv")
csv_filesPla.remove("./PlantCity/2023.csv")

csv_filesUma = glob.glob(os.path.join(pathUma, "*.csv"))
csv_filesUma.remove("./Umatilla/2021.csv")
csv_filesUma.remove("./Umatilla/2022.csv")
csv_filesUma.remove("./Umatilla/2023.csv")



csv_list = [csv_filesApo,
            csv_filesArc,
            csv_filesBal,
            csv_filesBro,
            csv_filesCit,
            csv_filesDov,
            csv_filesFlo,
            csv_filesLak,
            csv_filesPla,
            csv_filesUma]
     
print(csv_list)
```

```python
# Read each CSV file into DataFrame
import pandas as pd

big_df = pd.DataFrame()
dfs = []

for folder in csv_list:
    for file in folder:
        df = pd.read_csv(file)
        dfs.append(df)

big_df = pd.concat(dfs, ignore_index=True)
```

```python
big_df.columns = ['StationID', 'StationName', 'Date', 'Time', 'SoilTemp_C','Temp60cm_C', 'Temp2m_C','Temp10m_C', 'RelatHumidity','DewPointTemp_C', 'Rainfall_IN','WindSpeed_MPH', 'WindDirection_DEG','SolarRad_WM2']
```

```python
# Date: from object (Series) to date (YYYY-MM-DD)
# DOY: from date (YYYY-MM-DD) to day of year (doy = 1-366)

from datetime import datetime
import time

big_df['doy']=0
big_df['tod'] = 0

for index, row in big_df.iterrows():
    strDate = row['Date']
    objDate = datetime.strptime(strDate, "%Y-%m-%d")
    day = objDate.timetuple().tm_yday
    year = objDate.timetuple().tm_year
    mon = objDate.timetuple().tm_mon
    if (year == 2000 or year == 2004 or year == 2008 or year == 2012 or year == 2016 or year == 2020):
        if (mon > 2):
            day = day - 1
    
    big_df.at[index,'doy'] = day
    
    strTime = row['Time']
    objTime = time.strptime(strTime, "%H:%M:%S")
    big_df.at[index,'tod'] = objTime.tm_hour*4 + objTime.tm_min/15 + 1

```

```python
big_df['tod'] = big_df['tod'].astype(int)
big_df.dtypes
```

```python
sortedData = big_df.sort_values(by=['StationName','Date','Time'])
```

```python
# get data only after 2005
df = sortedData[~(sortedData['Date'] < '2005-01-01')]


for i in range(2005,2023):
    from_ts = str(i)+'-02-29'
    to_ts = str(i)+'-10-31'
    df = df[(df['Date'] < from_ts) | (df['Date'] > to_ts)]


#delete the start of 2005
from_ts = '2005-01-01'
to_ts = '2005-10-31'
df = df[(df['Date'] < from_ts) | (df['Date'] > to_ts)]

#delete the end of 2022
from_ts = '2022-03-01'
to_ts = '2022-12-31'
df = df[(df['Date'] < from_ts) | (df['Date'] > to_ts)]


df.head()
```
